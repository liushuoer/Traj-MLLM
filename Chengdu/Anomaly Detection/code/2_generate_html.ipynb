```python

import sys
from extract_traj_json import *
with open('../data/outlier_trajectories.pkl', 'rb') as f:
    trajectories = pickle.load(f)
```

```python
from tqdm import tqdm
trajectory_ids = list(trajectories.keys())
for trajectory_id in tqdm(trajectory_ids):
    trajectory = trajectories[trajectory_id]
    o_geo = traj2wgs(trajectory['o_geo'])
    o_geo = [[lon, lat] for lat, lon in o_geo]
    trajectory['o_geo'] = o_geo
    try:
        extract_full_traj_json(trajectory, trajectory_id, vis_type="poi")
    except Exception as e:
        print(f"Error processing trajectory ID {trajectory_id} (poi): {e}")
    try:
        extract_full_traj_json(trajectory, trajectory_id, vis_type="road_structure")
    except Exception as e:
        print(f"Error processing trajectory ID {trajectory_id} (road_structure): {e}")
```

```python
with open('../data/outlier_trajectories.pkl', 'rb') as f:
    trajectories = pickle.load(f)
```

```python
traj_id = list(trajectories.keys())[2]
trajectory = trajectories[traj_id]
o_geo = traj2wgs(trajectory['o_geo'])
o_geo = [[lon, lat] for lat, lon in o_geo]
trajectory['o_geo'] = o_geo
geo_points = trajectory['o_geo']

start_point = geo_points[0]
m = folium.Map(location=[start_point[1], start_point[0]], zoom_start=14)

folium.Marker(
    location=[geo_points[0][1], geo_points[0][0]],
    popup='Start',
    icon=folium.Icon(color='green', icon='play', prefix='fa')
).add_to(m)

folium.Marker(
    location=[geo_points[-1][1], geo_points[-1][0]],
    popup='End',
    icon=folium.Icon(color='red', icon='stop', prefix='fa')
).add_to(m)

# 绘制轨迹线
folium.PolyLine(
    locations=[[point[1], point[0]] for point in geo_points],
    color='blue',
    weight=3,
    opacity=0.8
).add_to(m)

folium.LayerControl().add_to(m)
title_html = f'''
    <h3 align="center" style="font-size:16px"><b>轨迹ID: {traj_id}</b></h3>
    <h4 align="center" style="font-size:14px">点位数量: {len(geo_points)}</h4>
'''
m.get_root().html.add_child(folium.Element(title_html))
```
