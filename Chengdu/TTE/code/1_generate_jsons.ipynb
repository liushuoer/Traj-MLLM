```python
import pickle
import sys
sys.path.append('//')
from extract_traj_json_chengdu import *
```

```python
with open('../../chengdu_demo.pkl', 'rb') as f:
    trajectories = pickle.load(f)
```

```python
from tqdm import tqdm

for trajectory in tqdm(trajectories):
    o_geo = traj2wgs(trajectory['o_geo'])
    o_geo = [[lon, lat] for lat, lon in o_geo]
    trajectory['o_geo'] = o_geo
    trajectory_id = trajectory['devid']

    try:
        extract_full_traj_json(trajectory, trajectory_id, vis_type="poi")
    except Exception as e:
        print(f"Error occurred while processing trajectory ID {trajectory_id} (poi): {e}")

    try:
        extract_full_traj_json(trajectory, trajectory_id, vis_type="road_structure")
    except Exception as e:
        print(f"Error occurred while processing trajectory ID {trajectory_id} (road_structure): {e}")
```
