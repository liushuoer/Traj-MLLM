```python
import os
import re
import shutil
import glob
import pickle
from tqdm import tqdm
import sys
from extract_traj_json_xian import *

source_dir = '../TTE/consolidated_data'
target_base_dir = '../TTE/consolidated_data'

os.makedirs(target_base_dir, exist_ok=True)

pattern = re.compile(r'trajectory_(.+?)(?:_\d+)?\.png')

all_files = os.listdir(source_dir)

traj_ids = set()
for file_name in all_files:
    match = pattern.match(file_name)
    if match:
        traj_ids.add(match.group(1))


with open('../xian_demo.pkl', 'rb') as f:
    trajectories = pickle.load(f)[:20]
```

```python
traj_ids = list(traj_ids)
```

```python
text_content = print_direction_and_speed_trend(trajectories[0], trajectories[0]["devid"])
text_content
```

```python
text_content
```

```python
# Process each trajectory ID
for trajectory in tqdm(trajectories):
    traj_id = trajectory["devid"]
    try:
        text_content = print_direction_and_speed_trend(trajectory, traj_id)
        txt_path = os.path.join(target_base_dir, f"consolidated_{traj_id}")
        print(f"consolidated_{traj_id}")
        txt_path = os.path.join(txt_path, f"dynamically_generated_stats_prompt_MP.txt")
        with open(txt_path, 'w', encoding='utf-8') as f:
            f.write(text_content)
    except Exception as e:
        print(f"Error occurred while processing trajectory ID {traj_id}: {e}")
print("Processing completed!")
```
