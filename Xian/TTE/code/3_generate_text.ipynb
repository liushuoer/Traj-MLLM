```python

import sys
from utils.get_prompt import generate_llm_prompt
sys.path.append('//')
from xian_extract_traj_json import *
with open('../../xian_demo.pkl', 'rb') as f:
    trajectories = pickle.load(f)
trajectories = trajectories[:20]
# edges_shp_path = 'edges.shp'
# edges_gdf = gpd.read_file(edges_shp_path)
# edges_dict_for_prompt_global = {row['fid']: row['highway'] for _, row in edges_gdf.iterrows()}
```

```python
from tqdm import tqdm

# Save dynamically_generated_stats_prompt as txt file
for trajectory in tqdm(trajectories):
    # if trajectory['devid']=="59088bdc6ce80a303730dabca27bb1e6":
    trajectory_id = trajectory['devid']
    segments = segment_trajectory(trajectory, trajectory_id, angle_threshold=30, merge_threshold=5)
    dynamically_generated_stats_prompt = generate_llm_prompt(
                        trajectory,
                        trajectory_id,  # devid
                        segments,
                        None,  # edges_dict
                        trajectories  # trajectories (complete dataset)
                    )
    timestamp_float = trajectory['tmp'][-1]
    datetime_object = datetime.fromtimestamp(timestamp_float).strftime('%Y-%m-%d %H:%M:%S')
    final_arrival_time = datetime.strptime(datetime_object, '%Y-%m-%d %H:%M:%S')
    print(final_arrival_time)
    dynamically_generated_stats_prompt = dynamically_generated_stats_prompt
    # target_dir, file_list = consolidate_images_for_trajectory(trajectory_id)
    # Save file_list as pickle file
    base_dir_1 = '../consolidated_data'
    target_consolidation_dir = os.path.join(base_dir_1, f"consolidated_{trajectory_id}")

    txt_file_path = os.path.join(target_consolidation_dir, 'dynamically_generated_stats_prompt.txt')
    if os.path.exists(txt_file_path):
        os.remove(txt_file_path)
    with open(txt_file_path, 'w', encoding='utf-8') as txt_file:
        txt_file.write(dynamically_generated_stats_prompt)

    # Save dynamically_generated_stats_prompt as txt file
    txt_file_path = os.path.join(target_consolidation_dir, 'ground_truth.txt')
    if os.path.exists(txt_file_path):
        os.remove(txt_file_path)
    with open(txt_file_path, 'w', encoding='utf-8') as txt_file:
        txt_file.write(str(final_arrival_time))
```
